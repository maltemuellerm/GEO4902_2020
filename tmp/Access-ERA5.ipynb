# Analyse the impact of the data assimilation - Surface Model

```python
import xarray as xr
```

##### ERA5 on thredds

```python
url='https://thredds.met.no/thredds/dodsC/metusers/maltem/WAM/ERA5_wave_Sep2018.nc'
ERA5 = xr.open_dataset(url)
```

### Find a specific point

```python
lat0=75; lon0=0;
point_era5 = ERA5.sel(lat=lat0,lon=lon0,method='nearest')
```

```python
from matplotlib import pyplot as plt

plt.plot(ERA5.time, point_era5.var229, '.-')

plt.title("Var229 - Sign. wave height")
plt.show()
```
